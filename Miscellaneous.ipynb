this file contain example of break,continue and matrix handiling

```python
#break statement
for i in range(6):
  if i==4: # stopes at 4
    break
  print(i)
```

```python
#continue statement
for i in range(6):
  if i==4: #skipped 4 and run rest
    continue
  print(i)
```
