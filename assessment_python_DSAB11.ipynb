Employee Attendance Bonus Eligibility
Scenario

A company provides a monthly attendance bonus to employees who meet the following criteria:

Attendance percentage must be 90% or above.
Late arrivals must not exceed 2.

Write a function that determines whether an employee is eligible for the attendance bonus.


```python

def check_bonus(attendance,late_count):
  if attendance >= 90 and late_count <= 2:
    return True
  else:
      return False
attendance = float(input("Attendance Percantage: "))
late_count = int(input("Late Arrivals: "))

result = check_bonus(attendance, late_count)

if attendance <= 90 and late_count <=2:
  print("Eligible for Bonus")
else:
  print("Not eligible for Bonus")
```

Challenge 2 (10 Marks)
Customer Feedback Analyzer
Scenario

An e-commerce company records customer ratings after each purchase.

The rating is converted into feedback labels.

Rating	Feedback
5	Excellent
4	Good
3	Average
1-2	Poor

The program should process multiple customer records and display the final feedback summary.



```python
def get_feedback(rating):
    if rating == 5:
        return "Excellent"
    elif rating == 4:
        return "Good"
    elif rating == 3:
        return "Average"
    elif rating == 1 or rating == 2:
        return "Poor"


def feedback_summary():
    n = int(input("Enter number of customers: "))

    feedback = {}

    for i in range(n):
        customer = input("Enter Customer Name: ")
        rating = int(input("Enter Rating: "))

        # Find feedback
        result = get_feedback(rating)

        # Store customer and feedback
        feedback[customer] = result

    print("\nCustomer Feedback Summary")

    # Display all customer feedback
    for customer, result in feedback.items():
        print(customer, ":", result)


feedback_summary()
```
